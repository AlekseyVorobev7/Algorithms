```python
def solve_17(rates):
    # сортируем по убыванию r — самые быстрорастущие покупаем раньше
    order = sorted(range(len(rates)), key=lambda i: -rates[i])
    total = sum(100 * (rates[order[t]] ** t) for t in range(len(rates)))
    return order, total

rates = [1.5, 1.2, 1.8, 1.3]
order, total = solve_17(rates)
print("Порядок покупки (индексы):", order)
print(f"Итоговая стоимость: {total:.2f}")
```

```python
def solve_17_2(rates):
    # r < 1; продаём сначала то, что падает медленнее (больший r)
    order = sorted(range(len(rates)), key=lambda i: -rates[i])
    total = sum(100 * (rates[order[t]] ** t) for t in range(len(rates)))
    return order, total

rates = [0.9, 0.7, 0.95, 0.8]
order, total = solve_17_2(rates)
print("Порядок продажи (индексы):", order)
print(f"Итоговая выручка: {total:.2f}")
```

```python
def is_subsequence(s_prime, s):
    i = 0  # указатель по s_prime
    for event in s:
        if i == len(s_prime):
            break
        if event == s_prime[i]:
            i += 1
    return i == len(s_prime)


S = ["Yahoo", "Yahoo", "eBay", "Yahoo", "Yahoo", "Oracle"]
S_prime = ["Yahoo", "eBay", "Yahoo", "Oracle"]
print(is_subsequence(S_prime, S))  # True

S2 = ["купить Amazon", "купить Yahoo", "купить eBay", "купить Yahoo", "купить Yahoo", "купить Oracle"]
S_prime2 = ["купить Yahoo", "купить eBay", "купить Yahoo", "купить Oracle"]
print(is_subsequence(S_prime2, S2))  # True

S_prime3 = ["купить Yahoo", "купить Oracle", "купить eBay"]
print(is_subsequence(S_prime3, S2))  # False — порядок нарушен
```
