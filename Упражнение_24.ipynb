```python
import networkx as nx
```

```python
G = nx.Graph()
G.add_edge(0, 1, weight=1)
G.add_edge(1, 2, weight=2)
G.add_edge(0, 2, weight=3)
G.add_edge(2, 3, weight=5)
G.add_edge(0, 3, weight=10)

# Находим MST T с помощью встроенного алгоритма Крускала
T = nx.minimum_spanning_tree(G, algorithm='kruskal')

print("Исходный граф G:")
print(list(G.edges(data=True)))
print("Исходное MST T:")
print(list(T.edges(data=True)))
```

```python
def find_max_weight_edge_on_path(T, new_u, new_v, new_weight):
    #Находит ребро с максимальным весом на уникальном пути
    #между вершинами u и v в дереве T. O(|V|) — поиск пути (DFS ИЛИ BFS) + проход по нему
    # Находим кратчайший путь по количеству ребер (в дереве он единственный)
    path = nx.shortest_path(T, source=new_u, target=new_v) # Тут BFS

    max_weight = -1
    max_edge = None

    # Проходим по всем ребрам на этом пути
    for i in range(len(path) - 1):
        node1 = path[i]
        node2 = path[i+1]

        # Получаем вес ребра
        weight = T[node1][node2]['weight']

        if weight > max_weight:
          max_weight = weight
          max_edge = (node1, node2)

    if new_weight < max_weight:
        print(f"Новое ребро легче ({new_weight} < {max_weight}) → T БОЛЬШЕ НЕ является MST.")
        return False, max_weight, max_edge
    else:
        print(f"Новое ребро не легче ({new_weight} >= {max_weight}) → T ОСТАЁТСЯ MST.")
        return True, max_weight, max_edge



```

```python
def update_mst(T, new_u, new_v, new_weight, edge_to_remove):
    #Пункт (b): Обновляет MST T, заменяя старое ребро на новое.
    #O(|V|) на поиск + O(1) на модификацию графа.

    # Создаем копию дерева, чтобы не менять оригинал
    new_T = T.copy()

    # Удаляем старое тяжелое ребро
    new_T.remove_edge(*edge_to_remove)
    print(f"Удалено ребро: {edge_to_remove}")

    # Добавляем новое, более легкое ребро
    new_T.add_edge(new_u, new_v, weight=new_weight)
    print(f"Добавлено ребро: ({new_u}, {new_v}, weight={new_weight})")
    return new_T
```

```python
# Параметры нового ребра: соединяем 0 и 3 с весом 3
# В текущем MST путь 0-1-2-3 имеет ребра весов 1, 2, 5. Максимум = 5.
# Так как 3 < 5, MST должно обновиться.
new_u, new_v, new_w = 0, 3, 3

is_still_mst, max_w_path, edge_to_remove = find_max_weight_edge_on_path(T, new_u, new_v, new_w)

if not is_still_mst:
    T_updated = update_mst(T, new_u, new_v, new_w, edge_to_remove)

    print(list(T_updated.edges(data=True)))
    total_weight = sum(d['weight'] for i, g, d in T_updated.edges(data=True))
    print(f"Общий вес нового MST: {total_weight}")
else:
    print("MST не изменилось.")
```

```python
#Параметры нового ребра: соединяем 0 и 3 с весом 6
# Максимум на пути все еще 5.
# Так как 6 > 5, MST должно остаться прежним.
new_u2, new_v2, new_w2 = 0, 3, 6

is_still_mst2, max_w_path2, edge_to_remove2 = find_max_weight_edge_on_path(T, new_u2, new_v2, new_w2)

if not is_still_mst2:
    print("\nВыполняем обновление MST...")
    T_updated2 = update_mst(T, new_u2, new_v2, new_w2, edge_to_remove2)
    print(f"Итоговое MST: {list(T_updated2.edges(data=True))}")
else:
    print("MST не изменилось")
    print(list(T.edges(data=True)))
```
